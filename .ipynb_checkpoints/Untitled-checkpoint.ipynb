```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("titanic/train.csv")
df
```

```python
%matplotlib notebook
plt.figure()
plt.bar([0,1], [len(df[df["Survived"]==0]),len(df[df["Survived"]==1])] , width=0.3)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

lab_enc=LabelEncoder()

y=df["Survived"]
dt=df.drop("Survived", axis=1)
feature_names=["Pclass", "Sex", "Age"]
X=dt[feature_names]
X["Sex"]=lab_enc.fit_transform(X["Sex"])

X["Age"]=X["Age"].fillna(method="backfill")
X
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y, random_state=0)
```

```python
logreg=LogisticRegression(C=20).fit(X_train,y_train)

y_pred_logistic=logreg.predict(X_test)
y_pred_logistic
```

```python
accuracy_score(y_test,y_pred_logistic)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier(n_estimators=10, max_depth=5).fit(X_train, y_train)

y_pred_randomforest=rfc.predict(X_test)
y_pred_randomforest
```

```python
accuracy_score(y_test,y_pred_randomforest)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier(max_depth=3).fit(X_train, y_train)

y_pred_decision_tree=rfc.predict(X_test)
y_pred_decision_tree
```

```python
accuracy_score(y_test,y_pred_decision_tree)
```

So using random forest classifier we will classify whether a passenger survives the Titanic sink.

```python
rfc_train=RandomForestClassifier().fit(X,y)
```

```python
df_test=pd.read_csv("titanic/test.csv")
df_test

feature_names=["Pclass", "Sex", "Age"]
X_test=df_test[feature_names]
X_test["Sex"]=lab_enc.fit_transform(X_test["Sex"])

X_test["Age"]=X_test["Age"].fillna(method="ffill")
X_test
```

```python
y_test_pred=rfc_train.predict(X_test)
y_test_pred
```

```python
plt.figure()
plt.bar([0,1], [len(y_test_pred[y_test_pred==0]),len(y_test_pred[y_test_pred==1])], width=0.3)
plt.show()
```

```python
df_test["Survived"]=y_test_pred
```

```python
col=["PassengerId","Survived"]
survival_log=df_test[col]
survival_log=survival_log.set_index("PassengerId")
survival_log
```
